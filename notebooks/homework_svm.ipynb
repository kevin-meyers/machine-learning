```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
breast_cancer = load_breast_cancer()
```

```python
data = pd.DataFrame(breast_cancer.data)
```

```python
breast_cancer.keys()
```

```python
data.columns = breast_cancer.feature_names
```

```python
targets = pd.Series(breast_cancer.target, name='target')
```

```python
targets
```

```python
sns.heatmap(data.corr())
```

```python
sns.heatmap(pd.DataFrame(data.corrwith(targets), columns=['target']))
```

```python
ss = StandardScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(ss.fit_transform(data), targets, test_size=0.25, random_state=0)
```

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
np.mean(svc.predict(X_test) == y_test)
```
